### Load MKA Dataset


```python
datasetPath = "./MKA datasets.zip"

excluded_classes = ["start", "menu", "pgdn", "pgup", "delete", "home", "neshanay xwarawa", "neshanay sarawa", "neshanay lay rast", "neshanay lay chap"]

mkaMatrixData = []
mkaLabels = []
mkaAudioData = []

with open(datasetPath, 'rb') as file:
    zip_data = BytesIO(file.read())

with zipfile.ZipFile(zip_data, 'r') as zip_file:
    file_list = zip_file.infolist()
    
    pattern = r"MKA datasets/(.*)/Sound Segment\(\.matrix\)/([^/]*)/.*\.Matrix"

    # loop through files in zip folder
    for file_info in zip_file.infolist():
        match = re.search(pattern, file_info.filename)
        if match:

            if match.group(2) in excluded_classes:
                continue

            print("Matching path:", file_info.filename)
            print("Extracted class:", match.group(2))

            mkaLabels.append(match.group(2))
            with zip_file.open(file_info.filename, "r") as file:
                mkaMatrixData.append({"label": match.group(2), "data": file.read()})

print(mkaMatrixData[500])
print(len(mkaMatrixData))
print(set(sorted(mkaLabels, key=len)))
```
